```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
import pandas as pd
import os
from time import time
from time import sleep
import seaborn as sns
import matplotlib.pyplot as plt

def api_runner():
    #df is what I updated everyday, it is the raw data.
    global df
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
      'start':'1',
      'limit':'5',
      'convert':'USD'
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': '484753e0-dd3b-4c21-90b4-060fd46451d5',
    }

    session = Session()
    session.headers.update(headers)

    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
      #print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)
    
    #This allows you to see all the columns and rows
    pd.set_option('display.max_column', None)
    pd.set_option('display.max_rows', None)
    #This normalizes the data and makes it all pretty in a dataframe
    df = pd.json_normalize(data['data'])
    df['timestamp'] = pd.to_datetime('now')
    #Put it into a csv file.
    if not os.path.isfile(r'/Users/shafanzheng/Downloads/DAProject/data_API'):#If the csv file does not exist, we creat a new one.
        df.to_csv(r'/Users/shafanzheng/Downloads/DAProject/data_API', header = 'column_names')
    else:
        df.to_csv(r'/Users/shafanzheng/Downloads/DAProject/data_API', mode = 'a', header = False)
    
for i in range(333):
    api_runner()
    print('Api runner completed successfully!')
    sleep(5) #sleep for 1 min
exit()
```

```python
#df2 is what I used to display. It is the updated csv file.
df2 = pd.read_csv(r'/Users/shafanzheng/Downloads/DAProject/data_API') 
df2
```

```python
#Data cleaning.
pd.set_option('display.float_format', lambda x: '%.2f' % x)
#df3 is certain columns that we need from df2.
df3 = df2.groupby('name',sort = False)[['quote.USD.percent_change_1h', 'quote.USD.percent_change_24h', 'quote.USD.percent_change_7d', 'quote.USD.percent_change_30d', 'quote.USD.percent_change_60d', 'quote.USD.percent_change_90d']].mean()
```

```python
#df4 is a more useable format of df3. It changes the data type of df3 from Dataframe to Series
df4 = df3.stack()
#df5 is a more useable format of df4. It changes the data type of df4 from Series to Dataframe but as the same format.
df5 = df4.to_frame(name = 'values')
df5
```

```python
df5.count()
```

```python
#df6 is df5 plus index number.
index = pd.Index(range(30))
df6 = df5.set_index(index)
df6 = df5.reset_index()
#df7 changed a column name of df6.
df7 = df6.rename(columns = {'level_1':'Period','values':'Percent_Change'})
df7['Period'] = df7['Period'].replace(['quote.USD.percent_change_1h', 'quote.USD.percent_change_24h', 'quote.USD.percent_change_7d', 'quote.USD.percent_change_30d', 'quote.USD.percent_change_60d', 'quote.USD.percent_change_90d'],['1h', '1d', '7d', '30d', '60d', '90d'])
```

```python
#Data visualization.
sns.catplot(x='Period', y='Percent_Change', hue='name', data=df7, kind='point')
#df3-df7 is for one usage.
```
